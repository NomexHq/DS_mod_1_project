```python
#second source is from imdb
from requests import get
from bs4 import BeautifulSoup
url = 'https://www.imdb.com/search/title?title_type=feature&release_date=2009-01-01,2018-12-31&sort=boxoffice_gross_us,desc&count=250'
response = get(url)
imdb_soup = BeautifulSoup(response.text, 'html.parser')
print(imdb_soup.prettify())
```

```python
# filtering to isolate all of the data that we want
imdb_movie_list = imdb_soup.find_all('div', class_ = 'lister-item mode-advanced')
# checking type and length
print(type(imdb_movie_list))
print(len(imdb_movie_list))
```

```python
# isolating one movie
movie1 = imdb_movie_list[0]
print(movie1.prettify())
```

```python
# isolating first movie's name
name1 = movie1.h3.a.text
name1
```

```python
#isolating first movie's year, removing parentheses and converting to integer
year1 = int(movie1.h3.find('span', class_='lister-item-year text-muted unbold').text.replace('(', '').replace(')', ''))
year1 
```

```python
# isoloating first movie's imdb score and converting to float
score1 = float(movie1.strong.text)
score1
```

```python
#isloating first movie's metacritic score and converting to int
mscore1 = int(movie1.find('span', class_='metascore favorable').text)
mscore1
```

```python
# isolating amount of votes on imdb and converting to int
numvotes1 = movie1.find('span', attrs = {'name': 'nv'})
numvotes1 = int(numvotes1.text.replace(',', ''))
numvotes1
```

```python
# isloating domestic gross
gross1 = int(movie1.find_all('span', attrs = {'name': 'nv'})[1]['data-value'].replace(',',''))
gross1
```

```python
# isolating first movie's genre and removing '\n'
genre1 = movie1.find('span', class_='genre')
genre1 = genre1.text.strip()
genre1
```

```python
# creating empty lists to scrape the data into
names = []
years = []
imdb_ratings = []
metascores = []
votes = []
genres = []
domestic = []
# looping over each 
for movie in imdb_movie_list:
    # name
    name = movie.h3.a.text
    names.append(name)
    # year
    year = movie.h3.find('span', class_ = 'lister-item-year text-muted unbold').text
#removing extra characters from some year strings     
    if '(I) ' in year:
        year = int(year.strip('(I)  '))
    elif '(II) ' in year:
        year = int(year.strip('(II)  '))
    elif '(IX) ' in year:
        year = int(year.strip('(IX)  '))
    else:
        year = int(year.strip('()'))
    years.append(year)
    # IMDB rating
    imdb = float(movie.strong.text)
    imdb_ratings.append(imdb)
    # Metascore
    m_score = movie.find('span', class_ = 'metascore').text
    metascores.append(int(m_score))
    # number of votes
    vote = movie.find('span', attrs = {'name':'nv'})['data-value']
    votes.append(int(vote))
    #domestic gross
    gross = int(movie1.find_all('span', attrs = {'name': 'nv'})[1]['data-value'].replace(',',''))
    domestic.append(gross)
    #genres
    genre = movie.find('span', class_='genre').text.strip()
    genres.append(genre)
```

```python
#testing data before looping over multiple pages
test = pd.DataFrame({'Title': names, 'Year': years, 'Genre': genres, 'IMDB_Score': imdb_ratings, 'Metascore': metascores, 'Votes': votes, 'US_Gross': domestic})
test.info()
```

```python
# creating variable to edit url to loop 
from random import randint
pages = [str(i) for i in range(1, 752, 250)]
```

```python
#big loop!
names = []
years = []
imdb_ratings = []
metascores = []
votes = []
genres = []
domestic = []

for page in pages:
    response = get('https://www.imdb.com/search/title?title_type=feature&release_date=2009-01-01,2018-12-31&sort=boxoffice_gross_us,desc&count=250&start=' + page)
    html = BeautifulSoup(response.text, 'html.parser')
    imdb_movie_list = html.find_all('div', class_='lister-item mode-advanced')
    for movie in imdb_movie_list:
        # name
        name = movie.h3.a.text
        names.append(name)
        # year
        year = movie.h3.find('span', class_ = 'lister-item-year text-muted unbold').text  
        if '(I) ' in year:
            year = int(year.strip('(I)  '))
        elif '(II) ' in year:
            year = int(year.strip('(II)  '))
        elif '(III) ' in year:
            year = int(year.strip('(III)  '))
        elif '(VI) ' in year:
            year = int(year.strip('(VI)  '))   
        elif '(IX) ' in year:
            year = int(year.strip('(IX)  '))
        else:
            year = int(year.strip('()'))
        years.append(year)
        # IMDB rating
        imdb = float(movie.strong.text)
        imdb_ratings.append(imdb)
        # Metascore - need to account for null values
        m_score = movie.find('span', class_ = 'metascore')
        if m_score is None:
            m_score = '?'
        else:
            m_score = m_score.text
        metascores.append(m_score)
        # number of votes
        vote = movie.find('span', attrs = {'name':'nv'})['data-value']
        votes.append(int(vote))
        #domestic gross
        gross = int(movie1.find_all('span', attrs = {'name': 'nv'})[1]['data-value'].replace(',',''))
        domestic.append(gross)
        #genres
        genre = movie.find('span', class_='genre').text.strip()
        genres.append(genre)
```

```python
# creating and saving dataframe
imdb_df = pd.DataFrame({'Title': names, 'Year': years, 'Genre': genres, 'IMDB_Score': imdb_ratings, 'Metascore': metascores, 'Votes': votes, 'US_Gross': domestic})
imdb_df.info()
imdb_df.to_csv('imdb.csv')
```
