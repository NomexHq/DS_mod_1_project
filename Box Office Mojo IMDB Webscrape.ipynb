```python
from requests import get
from bs4 import BeautifulSoup
url = 'https://www.imdb.com/search/title?title_type=feature&release_date=2009-01-01,2018-12-31&sort=boxoffice_gross_us,desc&count=250'
response = get(url)
imdb_soup = BeautifulSoup(response.text, 'html.parser')
type(imdb_soup)
```

```python
movie_containers = imdb_soup.find_all('div', class_ = 'lister-item mode-advanced')
print(type(movie_containers))
print(len(movie_containers))
```

```python
first_movie = movie_containers[0]
first_movie
```

```python
first_name = movie_containers[0].h3.a.text
first_name
```

```python
first_year = first_movie.h3.find('span', class_='lister-item-year text-muted unbold')
first_year = first_year.text
```

```python
first_score = float(first_movie.strong.text)
first_score
```

```python
first_mscore = int(first_movie.find('span', class_='metascore favorable').text)
first_mscore
```

```python
first_numvotes = first_movie.find('span', attrs = {'name': 'nv'})
first_numvotes = int(first_numvotes.text.replace(',', ''))
first_numvotes
```

```python
first_genre = first_movie.find('span', class_='genre')
first_genre = first_genre.text.strip()
```

```python
# Lists to store the scraped data in
names = []
years = []
imdb_ratings = []
metascores = []
votes = []
genres = []
for container in movie_containers:
    
        # The name
        name = container.h3.a.text
        names.append(name)
        # The year
        year = container.h3.find('span', class_ = 'lister-item-year').text
        years.append(year)
        # The IMDB rating
        imdb = float(container.strong.text)
        imdb_ratings.append(imdb)
        # The Metascore
        m_score = container.find('span', class_ = 'metascore').text
        metascores.append(int(m_score))
        # The number of votes
        vote = container.find('span', attrs = {'name':'nv'})['data-value']
        votes.append(int(vote))
        #The genres
        genre = container.find('span', class_='genre').text.strip()
        genres.append(genre)
```
