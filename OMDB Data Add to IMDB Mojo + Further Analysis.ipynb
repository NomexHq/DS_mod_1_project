```python
# filtering for movies that made at least 500MM
money = movies[movies['Worldwide_Gross'] >= 500]
```

```python
# examing most common words in plots and removing stopwords
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
set(stopwords.words('english'))
```

```python
money.info()
```

```python
stop = stopwords.words('english')
money_plots = money['Plot']
money_plots.dropna(inplace=True)
```

```python
stop
```

```python
money_plots.reset_index(drop=True)
```

```python
# removing stop words and other words not included in the list
words = money_plots.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
words = words.str.replace('The', '')
words = words.str.replace('must', '')
words = words.str.replace('take', '')
words = words.str.replace('new', '')
words
```

```python
movies.columns
```

```python
# putting new list of words into a wordcloud
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
%matplotlib inline
text = " ".join(descr for descr in words)
```

```python
# Create and generate a word cloud image:
wordcloud = WordCloud(max_font_size=250, max_words=100, height=800, width=1600, background_color='white').generate(text)
# Display the generated image:
plt.figure(figsize=(20,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
wordcloud.to_file('worldcloud.png')
```

```python
# looking at movies that made both at least 500mm and scored over a 60 metascore - metacritic defines minimum score of generally positive at 61
top_movies = movies.loc[:, ('Title','Metascore', 'Worldwide_Gross', 'Director')].sort_values(by='Metascore', ascending=False)
best_movies = top_movies[top_movies['Worldwide_Gross']>=500]
best_movies = best_movies[best_movies['Metascore']>60]
best_movies['Director'].value_counts().head(7)
```

```python
# filtering data using list of best directors
best_directors= best_movies[(best_movies['Director'] == 'Christopher Nolan') | (best_movies['Director'] == 'Francis Lawrence') | 
            (best_movies['Director'] == 'Anthony Russo, Joe Russo') | (best_movies['Director'] == 'James Gunn') | 
            (best_movies['Director'] == 'Brad Bird') | (best_movies['Director'] == 'Christopher McQuarrie ') | 
            (best_movies['Director'] == 'Pierre Coffin, Chris Renaud')]
best_directors
```
